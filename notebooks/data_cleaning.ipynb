# Automated Data Cleaning Pipeline

This notebook demonstrates how to execute the end-to-end data cleaning pipeline we built in the `src/` directory.

```python
import sys
import os
import pandas as pd
import numpy as np

# Add src to path so we can import our modules
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))

from pipelines.data_cleaning_pipeline import DataCleaningPipeline
```

```python
# Initialize the pipeline
pipeline = DataCleaningPipeline(raw_data_dir='../data/raw', processed_data_dir='../data/processed')

# Create a dummy raw dataset for demonstration
dummy_data = {
    'id': [1, 2, 2, 4, 5],
    'name': [' Alice ', 'Bob', 'Bob', None, 'EVE'],
    'age': [25, 30, 30, 150, np.nan], # 150 is an outlier
    'date': ['2023-01-01', '02/01/2023', '02/01/2023', 'invalid', '2023-01-05']
}
df_raw = pd.DataFrame(dummy_data)
df_raw.to_csv('../data/raw/dummy_raw.csv', index=False)
print("Raw Data Saved.\n")

# Run the pipeline!
df_clean = pipeline.run_pipeline(
    filename='dummy_raw.csv',
    date_columns=['date'],
    scale_method='standard',
    output_filename='dummy_clean.csv'
)

print("\n--- Final Cleaned Dataset ---")
display(df_clean)
```
